**Read Checkins data**

```python
check_in = pd.read_csv("../datasets/dataset_WWW2019/dataset_WWW2019/dataset_WWW_Checkins_anonymized.txt", sep="\t", header=None)
```

```python
check_in.columns = ['user_id','venue_id','utc_time', 'offset']
```

```python
check_in.head()
```

```python
check_in.shape
```

**Read venue data**

```python
venue = pd.read_csv("../datasets/dataset_WWW2019/dataset_WWW2019/raw_POIs.txt", sep="\t", header=None)
```

```python
venue.columns = ['venue_id', 'latitude', 'longitude', 'category', 'country_code']
```

```python
venue.head()
```

```python
venue.shape
```

```python
# rounding off the latitude and longitude decimals upto 5 
venue['longitude'] = venue['longitude'].apply(lambda x: round(x, 5))
venue['latitude'] = venue['latitude'].apply(lambda x: round(x, 5))
```

```python
# considering only New York data
venue = venue[ (venue['longitude'] >= -74.2748) & (venue['longitude'] <= -73.70741) & \
              (venue['latitude'] >= 40.45326) & (venue['latitude'] <= 40.94788)]
```

```python
venue.shape
```

**Merging checkins with venue data**

```python
network_df = pd.merge(check_in, venue, on="venue_id")
network_df.head()
```

```python
network_df.shape
```

**Reading friend relationship data**

```python
friends = pd.read_csv("../datasets/dataset_WWW2019/dataset_WWW2019/dataset_WWW_friendship_new.txt", sep="\t", header=None)
```

```python
friends.columns = ['user_id','user2']
```

```python
friends.head()
```

```python
# friends.to_csv("../datasets/friends.csv", encoding='utf-8')
```

**Merging friend relationship data with actual checkin data**

```python
net_df = pd.merge(network_df, friends, how='inner', on= "user_id")
```

```python
net_df.shape
```

```python
l_user_id = list(net_df['user_id'].unique())
l_user_id
```

```python
net_df = net_df[net_df['user2'].isin(l_user_id)]
```

```python
net_df.shape
```

```python
net_df.to_csv("../datasets/network_df_new_inner.csv",encoding='utf-8')
```
